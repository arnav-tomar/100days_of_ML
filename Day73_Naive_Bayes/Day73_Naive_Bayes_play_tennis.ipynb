## 1. Imports

```
import pandas as pd
import numpy as np
```

## 2. Dataset

```
data = pd.DataFrame({
    'outlook': ['Sunny','Sunny','Overcast','Rain','Rain','Rain','Overcast','Sunny','Sunny','Rain','Sunny','Overcast','Overcast','Rain'],
    'temp': ['Hot','Hot','Hot','Mild','Cool','Cool','Cool','Mild','Cool','Mild','Mild','Mild','Hot','Mild'],
    'humidity': ['High','High','High','High','Normal','Normal','Normal','High','Normal','Normal','Normal','High','Normal','High'],
    'wind': ['Weak','Strong','Weak','Weak','Weak','Strong','Strong','Weak','Weak','Weak','Strong','Strong','Weak','Strong'],
    'play': ['No','No','Yes','Yes','Yes','No','Yes','No','Yes','Yes','Yes','Yes','Yes','No']
})
data
```

## 3. Priors

```
counts = data['play'].value_counts()
py = counts['Yes'] / len(data)
pn = counts['No'] / len(data)
print('P(Yes)=', py)
print('P(No)=', pn)
```

## 4. Frequency Tables

```
print(pd.crosstab(data['outlook'], data['play']))
print(pd.crosstab(data['temp'], data['play']))
print(pd.crosstab(data['humidity'], data['play']))
print(pd.crosstab(data['wind'], data['play']))
```

## 5. Manual Likelihoods

```
P_sunny_yes, P_sunny_no = 2/9, 3/5
P_hot_yes, P_hot_no = 2/9, 2/5
P_high_yes, P_high_no = 3/9, 4/5
P_weak_yes, P_weak_no = 6/9, 2/5
```

## 6. Prediction: (Sunny, Hot, High, Weak)

```
score_yes = py * P_sunny_yes * P_hot_yes * P_high_yes * P_weak_yes
score_no = pn * P_sunny_no * P_hot_no * P_high_no * P_weak_no

print('Score Yes =', score_yes)
print('Score No =', score_no)
print('Prediction =', 'Yes' if score_yes > score_no else 'No')
```
