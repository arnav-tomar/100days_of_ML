# Working with dates

```python
# Converting to dotetime datatype
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

#1. Extract year

```python
date['d_year'] = date['date'].dt.year

date.head()
```

```python
date['d_month'] = date['date'].dt.month
```

```python
date['d_day'] = date['date'].dt.day
```

```python
date.head()
```

```python
date['d_m_name'] = date['date'].dt.month_name()
```

```python
date['d_d_owd'] = date['date'].dt.dayofweek
```

```python
date.head(15)
```

```python
date['d_d_ow'] = date['date'].dt.day_name()
```

```python
date.head(15)
```

```python
# is weekend?

date['date_is_weekend'] = np.where(date['d_d_ow'].isin(['Sunday', 'Saturday']), 1,0)

date.drop(columns=['product_id','city_id','orders']).head(15)
```

```python
date['date_week'] = date['date'].dt.isocalendar().week
date = date.drop(columns=['product_id', 'city_id', 'orders'])
date.head()
```

```python
date['quarter'] = date['date'].dt.quarter
date.head()
```

```python
date['semester'] = np.where(date['quarter'].isin([1,2]), 1, 2)
```

```python
date.head()
```

```python
# Extract Time elapsed between dates

import datetime

today = datetime.datetime.today()

today
```

```python
(today - date['date'])
```

```python
(today - date['date']).dt.days
```

```python
# Calculate months passed
# np.timedelta64(1, 'M') is not supported due to variable month lengths.
# A more robust way to calculate months between dates:
months_passed = ((today.year - date['date'].dt.year) * 12 + (today.month - date['date'].dt.month))
print(months_passed)
```
