```python
import pandas as pd
import numpy as np

years = range(2014, 2024)
months = range(1, 13)

data = []
for year in years:
    for month in months:
        data.append({'year': year, 'month': month, 'passenger': np.random.randint(100, 1000)})

passenger_df = pd.DataFrame(data)

display(passenger_df)
```

```python
sns.heatmap(passenger_df.pivot_table(values='passenger', index='month', columns='year'))
```
