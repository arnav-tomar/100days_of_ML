# üß± 1Ô∏è‚É£ Decision Boundary (Hyperplane)

## Hyperplane Equation
$$ w^T x + b = 0 $$

### Symbols
| Symbol | Meaning |
|--------|--------|
| $w$ | Weight vector |
| $x$ | Feature vector |
| $b$ | Bias |
| $w^T x$ | Dot product |

This is the **decision boundary**.

---

# üìê 2Ô∏è‚É£ Margin Planes

Three parallel planes:

### Decision Boundary
$$ w^T x + b = 0 $$

### Upper Margin
$$ w^T x + b = +1 $$

### Lower Margin
$$ w^T x + b = -1 $$

These define the **margin band**.

---

# ‚≠ê 3Ô∏è‚É£ Support Vectors

Points exactly on margins.

## Condition
$$ y_i (w^T x_i + b) = 1 $$

They:
- Define hyperplane  
- Control margin  
- Are critical points  

Remove them ‚Üí boundary changes.

---

# üìè 4Ô∏è‚É£ Margin Derivation

Distance between parallel planes.

---

## Distance Between Two Parallel Planes

General formula:
$$
\text{Distance} = \frac{|b_2 - b_1|}{\|w\|}
$$

Planes:
- $w^T x + b = +1$  
- $w^T x + b = -1$  

---

## Final Margin Formula

$$
\text{Margin} = \frac{2}{\|w\|}
$$

‚≠ê Most important result.

---

# üéØ 5Ô∏è‚É£ Optimization Objective

We want maximum margin:

$$
\max \frac{2}{\|w\|}
$$

Equivalent to:

$$
\min \|w\|
$$

Common convex form:

$$
\min \|w\|^2
$$

---

# üìå 6Ô∏è‚É£ Hard Margin Constraint

All data must satisfy:

$$
y_i (w^T x_i + b) \ge 1
$$

Ensures:
- Correct classification  
- No violations  

This defines **Hard Margin SVM**.

---

# üß† 7Ô∏è‚É£ Why $\pm 1$?

Scaling trick.

Hyperplane scaling doesn't change boundary:

$$
w^T x + b = 0 \quad \equiv \quad k(w^T x + b) = 0
$$

So we normalize margins to:
$$ \pm 1 $$

Called **canonical form**.

---

# üìê 8Ô∏è‚É£ Geometry of $w$

Key insights:

- $w \perp$ hyperplane  
- Larger $\|w\|$ ‚Üí smaller margin  
- Smaller $\|w\|$ ‚Üí larger margin  

Thus:

$$
\min \|w\| \Rightarrow \max \text{margin}
$$

---

# üìå 9Ô∏è‚É£ Final Optimization (Primal Form)

### Objective
$$
\min \frac{1}{2} \|w\|^2
$$

### Subject to
$$
y_i (w^T x_i + b) \ge 1
$$

This is the **standard Hard Margin SVM formulation**.

---

# üìè üîü Distance of Point from Hyperplane

Distance formula:

$$
\frac{|w^T x + b|}{\|w\|}
$$

Used for:
- Margin derivation  
- Confidence scoring  
- Support vector detection  

---

# üéØ Core Intuition

Choose hyperplane maximizing minimum distance:

$$
\max \min_i \frac{|w^T x_i + b|}{\|w\|}
$$

Classic **max‚Äìmin optimization**.

---

# ‚ùó Limitation of Hard Margin

Works only if data is perfectly separable:

$$
\exists (w, b) \text{ such that } y_i (w^T x_i + b) > 0
$$

Fails with:
- Noise  
- Overlap  

Solution ‚Üí Soft Margin SVM.

---

# üßæ Formula Summary

### Hyperplane
$$ w^T x + b = 0 $$

### Margins
$$ w^T x + b = \pm 1 $$

### Support vectors
$$ y_i (w^T x_i + b) = 1 $$

### Margin width
$$ \frac{2}{\|w\|} $$

### Objective
$$ \min \|w\|^2 $$

### Constraint
$$ y_i (w^T x_i + b) \ge 1 $$

---

# üß† Derivation Flow

1. Draw separating hyperplane  
2. Add parallel margin planes  
3. Identify support vectors  
4. Compute plane distance  
5. Convert max margin ‚Üí min norm  
6. Add classification constraint  
7. Final convex optimization  

---

# üéì Interview Gold

### Why minimize $\|w\|$?
Because:
$$
\text{Margin} = \frac{2}{\|w\|}
$$

---

### Why squared norm?
Convex + differentiable optimization.

---

### Why support vectors matter?
They alone determine boundary.

---

### Why normalize to $\pm 1$?
Scaling invariance of hyperplane.

# ‚öñÔ∏è 4Ô∏è‚É£ Classification Condition

Basic linear classifier:

$$

Classification rule:

- $> 0$ ‚Üí positive class  
- $< 0$ ‚Üí negative class

# üìè 5Ô∏è‚É£ Distance-Based Thinking

Speaker repeatedly emphasizes distance.

Distance from hyperplane:

$$

# üß† 6Ô∏è‚É£ Margin Interpretation

Important intuition:

- Larger margin ‚Üí better generalization  
- Smaller margin ‚Üí overfitting risk  

So:

$$
\text{Generalization} \propto \text{Margin}

# üìâ 7Ô∏è‚É£ Margin vs Error Tradeoff

Implicit idea explained:

If we force perfect separation:

$$
\text{Margin} \downarrow
$$

If we allow small errors:

$$
\text{Margin} \uparrow
$$

This leads toward **Soft Margin SVM**.

---

# ‚öôÔ∏è 8Ô∏è‚É£ Optimization Thinking

Lecture repeatedly hints:

We minimize something.

Core objective:

$$
\min \|w\|^2
$$

Why squared?

- Smooth gradients  
- Convex optimization  
- Easier calculus  

---

# üß™ 9Ô∏è‚É£ Imperfect Classification Reality

Important real-world statement:

Some points may lie:

- Inside margin  
- On wrong side  

So constraint becomes relaxed.

Hard constraint:

$$
y_i (w^T x_i + b) \ge 1
$$

Real-world version:

$$
y_i (w^T x_i + b) \ge 1 - \xi_i
$$

Where:

$$
\xi_i = \text{slack variable}
$$

---

# üß© üîü Slack Variables (Soft Margin Intro)

Slack variable meaning:

$$
\xi_i \ge 0
$$

Cases:

| $\xi_i$ value | Meaning |
|--------------|--------|
| $0$ | Correct + outside margin |
| $0 < \xi_i < 1$ | Inside margin |
| $\xi_i > 1$ | Misclassified |

---

# üéØ 1Ô∏è‚É£1Ô∏è‚É£ New Optimization Objective

Now we balance:

- Margin maximization  
- Error minimization  

So objective becomes:

$$
\min \frac{1}{2} \|w\|^2 + C \sum \xi_i
$$

Where:

$$
C = \text{regularization parameter}
$$

---

# ‚öñÔ∏è 1Ô∏è‚É£2Ô∏è‚É£ Meaning of $C$

Trade-off controller.

### Large $C$
$$
\text{Low error, small margin}
$$

### Small $C$
$$
\text{Large margin, more tolerance}
$$

---

# üß† 1Ô∏è‚É£3Ô∏è‚É£ Intuition Summary from Lecture

Speaker repeatedly reinforces:

- Real world has noise  
- Perfect lines unrealistic  
- Allow flexibility  
- Maximize margin responsibly  

This is the philosophical jump:

Hard Margin ‚Üí Soft Margin.

# ‚ùó 1Ô∏è‚É£6Ô∏è‚É£ Practical Interpretation

Lecture analogy repeated:

"Reduce distance ‚Üí bad boundary  
Increase distance ‚Üí strong boundary"

So we aim to:

$$

Classic max‚Äìmin optimization.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.axes._axes import _log as matplotlib_axes_logger
from mpl_toolkits import mplot3d
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from matplotlib.colors import ListedColormap
```

```python
from sklearn.datasets import make_circles
X, y = make_circles(500, factor=.15, noise=.15)

plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='bwr')
```

```python
# linear
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20)
classifier = SVC(kernel="linear")
classifier.fit(X_train, y_train.ravel())
y_pred = classifier.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

# visual + x**2 on r

```python
def plot_3d_plot(X, y):
    r = np.exp(-(X ** 2).sum(1))
    ax = plt.subplot(projection='3d')
    ax.scatter3D(X[:, 0], X[:, 1], r, c=y, s=100, cmap='bwr')
    ax.set_xlabel('X1')
    ax.set_ylabel('X2')
    ax.set_zlabel('y')
    return ax
plot_3d_plot(X,y)
```

```python
rbf_classifier = SVC(kernel="rbf")
rbf_classifier.fit(X_train, y_train)
y_pred = rbf_classifier.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
def plot_decision_boundary(X, y, classifier):
    plt.figure(figsize=(8, 6))

    # Create a mesh to plot in
    x_min, x_max = X[:, 0].min() - 0.1, X[:, 0].max() + 0.1
    y_min, y_max = X[:, 1].min() - 0.1, X[:, 1].max() + 0.1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100),
                           np.linspace(y_min, y_max, 100))

    # Predict the classification for each point in the mesh
    Z = classifier.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # Plot the decision boundary
    plt.contourf(xx, yy, Z, alpha=0.4, cmap='bwr')
    plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='bwr', edgecolors='k')

    plt.title('SVM Decision Boundary')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.show()
```

```python
plot_decision_boundary(X, y, rbf_classifier)
```

```python
poly_classifier = SVC(kernel="poly", degree=2)
poly_classifier.fit(X_train, y_train)
y_pred = poly_classifier.predict(X_test)
```

```python
plot_decision_boundary(X, y, poly_classifier)
```


# üß† Mathematical Transformations
# üéØ 1Ô∏è‚É£ Core Idea of Transformation

A transformation means:

$$
x \rightarrow T(x)
$$

Where:
- $x$ = original representation  
- $T(x)$ = transformed representation  

Goal:
Make problem easier.

---

# üß† 2Ô∏è‚É£ Why Transform Data?

Main motivation:

> Hard problems become easy after transformation.

Mathematically:

$$
\text{Solve in new space instead of original space}
$$

---

# üìê 3Ô∏è‚É£ Feature Transformation

Convert input features:

$$
x = (x_1, x_2) \rightarrow \phi(x)
$$

Example:

$$
\phi(x) = (x_1^2, x_2^2, x_1 x_2)
$$

Now data lives in:

Higher dimensional space.

---

# üåç 4Ô∏è‚É£ Real World Meaning

Speaker intuition:

- Raw data = messy  
- Transformed data = structured  

So:

$$
\text{Representation} \Rightarrow \text{Performance}
$$

Better features ‚Üí better models.

---

# ‚öôÔ∏è 5Ô∏è‚É£ Transformations in Classification

Original classifier:

$$
f(x) = w^T x + b
$$

After transformation:

$$
f(x) = w^T \phi(x) + b
$$

Same linear model ‚Äî
But in **new feature space**.

---

# üß† 6Ô∏è‚É£ Linear vs Nonlinear Problems

Key idea:

Some problems are nonlinear in original space:

$$
\text{Nonlinear in } x
$$

But become linear after transform:

$$
\text{Linear in } \phi(x)
$$

This is core ML trick.

---

# üìä 7Ô∏è‚É£ Example Intuition

Circle classification example.

Original boundary:

$$
x_1^2 + x_2^2 = r^2
$$

Nonlinear in 2D.

Transform:

$$
z = x_1^2 + x_2^2
$$

Now boundary becomes:

$$
z = r^2
$$

Linear in transformed space.

---

# üîÑ 8Ô∏è‚É£ Transformation as Mapping

General mapping:

$$
\phi : \mathbb{R}^n \rightarrow \mathbb{R}^m
$$

Where:

- $n$ = original dimension  
- $m$ = transformed dimension  

Often:

$$
m > n
$$
